```python
import csv
import pandas as pd
import numpy as np
from datetime import datetime
import bar_chart_race as bcr
```

```python
df = pd.read_csv('content/001.tablaUsos.csv')
df = df[df['Sede']=='Arv']
df['Fecha']=pd.to_datetime(df['Fecha'])
df['Total'] = df['Total'].astype('float64')
df['Created By'] = df['Created By'].astype(str)
df.head()
df.columns
```

```python
df = df[['Fecha', 'Created By', 'Total']]
df.info()
```

```python
df.head()
```

```python
QryAllUsers = df.pivot_table(index='Fecha', columns= 'Created By', aggfunc = {'Total':'sum'})['Total'].fillna(0)
QryAllUsers.head()
```

```python
QryAllUsers = QryAllUsers.rename(columns={
    'Antonio Alexander Henao Marulanda': 'Too',
       'Janet Villa Galvis': 'Janet',
       'Jhomaro Alberto Gaviria Catao': 'Jhoma', 'Juliana Snchez Serna': 'Juli',
       'Laura Sofia Vinasco Gutierrez': 'Sofi', 'Luz Ceneida Alzate Rodriguez': 'Cene',
       'Vivian Vsquez Parra': 'ViviVasquez'
})
QryAllUsers.head()
```

```python
#Esta funcin permite acumular las cantidades en cada fecha:

def features_acum(data):
  data_acum= pd.DataFrame(None)
  for c in data.columns:
    y=[]
    x = data[c].values
    for i in range(len(x)):
      if (i -1)<0:
        y.append(x[i])
      else:
        y.append(x[i]+y[i-1])
    data[f' {c}']=y # aqu estoy guardando el nombre
    del(data[c])
  return data
```

```python
def create_bar_chart_race(data, description):
  bcr_html = bcr.bar_chart_race(
      df=data,
      filename=f"{description}.gif", #'elnombreaqui.mp4' cuando requiera una copia del ejemplar
      orientation='h',
      sort='desc',
      n_bars=10, #nmero de caracteristicas para mostrar
      fixed_order=False,
      fixed_max=True,
      steps_per_period=25,
      interpolate_period=False,
      label_bars=True,
      bar_size=.95,
      period_label={'x': .99, 'y': .35, 'ha': 'right', 'va': 'center'},
      #period_fmt= '%M %d, %Y',
      period_summary_func=lambda v, r: {'x': .99, 'y': .18,
                                        's': f'Total usos : {v.nlargest(6).sum():,.0f}\nFuente: Estadstica, 001.tablaUsos',
                                        'ha': 'right', 'size': 8} ,
      #perpendicular_bar_func='median',
      period_length=600, #Mientras ms pequeo mayor velocidad y viceversa
      figsize=(5, 3),
      dpi=144,
      cmap='PiYG', #por defecto 'dark12', cambia el color de las bar de cada caracteristica opciones : https://matplotlib.org/3.1.1/tutorials/colors/colormaps.html
      title= description,
      title_size=12,
      bar_label_size=11,
      tick_label_size=13,
    #   shared_fontdict={'family' : 'Arial', 'color' : '.3'},
    #   shared_fontdict={'color' : '.3'},
      scale='linear',
      writer='ffmpeg',
      fig=None,
      bar_kwargs={'alpha': .7},
      filter_column_colors=True)  #Por defecto False
  return bcr_html
```

```python
QryAllUsersAcum = features_acum(QryAllUsers) # tener cuidado con esta ejecucin, hgalo solo una vez para evitar dobles acumulados
QryAllUsersAcum.head()
```
