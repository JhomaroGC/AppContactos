```python
import csv
import pandas as pd
import numpy as np
from datetime import datetime
import bar_chart_race as bcr
```

```python
df = pd.read_csv('content\\024.estadoPedidosBodega.csv')

df['Fecha'] = df['Created'].str.split(' ', expand=True)[0]
df['Fecha'] = pd.to_datetime(df['Fecha'], format='%d/%m/%Y')
df['cantidad'] = df['cantidad'].astype('float64')
df['Nombre'] = df['solicitante'].str.split(' ', expand=True)[0]+df['solicitante'].str.split(' ', expand=True)[1]
df['Created By'] = df['Created By'].astype(str)
display(df.columns)
df = df[['Fecha', 'Producto', 'cantidad']]
```

```python
#Esta función permite acumular las cantidades en cada fecha:

def features_acum(data):
  data_acum= pd.DataFrame(None)
  for c in data.columns:
    y=[]
    x = data[c].values
    for i in range(len(x)):
      if (i -1)<0:
        y.append(x[i])
      else:
        y.append(x[i]+y[i-1])
    data[f'➤ {c}']=y # aquí estoy guardando el nombre
    del(data[c])
  return data
```

```python
def create_bar_chart_race(data, description):
  bcr_html = bcr.bar_chart_race(
      df=data,
      filename=f"{description}.gif", #'elnombreaqui.mp4' cuando requiera una copia del ejemplar
      writer='ffmpeg',
      orientation='h',
      sort='desc',
      n_bars=10, #número de caracteristicas para mostrar
      fixed_order=False,
      fixed_max=True,
      steps_per_period=25,
      interpolate_period=False,
      label_bars=True,
      bar_size=.95,
      period_label={'x': .99, 'y': .35, 'ha': 'right', 'va': 'center'},
      #period_fmt= '%M %d, %Y',
      period_summary_func=lambda v, r: {'x': .99, 'y': .18,
                                        's': f'Total Productos Solicitados en Bodega: {v.nlargest(6).sum():,.0f}\nFuente: equipoArví: 024.estadoPedidosBodega',
                                        'ha': 'right', 'size': 8} ,
      #perpendicular_bar_func='median',
      period_length=600, #Mientras más pequeño mayor velocidad y viceversa
      figsize=(5, 3),
      dpi=144,
      cmap='dark12', #por defecto 'dark12', cambia el color de las bar de cada caracteristica opciones : https://matplotlib.org/3.1.1/tutorials/colors/colormaps.html, PiYG: magentica
      title= description,
      title_size=12,
      bar_label_size=11,
      tick_label_size=8,
    #   shared_fontdict={'family' : 'Arial', 'color' : '.3'},
    #   shared_fontdict={'color' : '.3'},
      scale='linear',
      # writer=None
      fig=None,
      bar_kwargs={'alpha': .7},
      filter_column_colors=True)  #Por defecto False
  return bcr_html
```
